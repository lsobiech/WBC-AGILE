```python
import pandas as pd
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers
import shutil

# Ensure the base target directory exists
base_images_dir = 'Images'
os.makedirs(base_images_dir, exist_ok=True)

# Unpack the archive directly into the base_images_dir
shutil.unpack_archive('part1/Images.zip', base_images_dir)

# Adjust images_dir to point to the actual folder containing images inside the unpacked archive
images_dir = os.path.join(base_images_dir, 'Images') # Assuming the zip contains a top-level 'Images' folder

labels_df = pd.read_csv('part1/labels.csv', header=None)
labels_df_columns = ['image_id','class','x_min','y_min','x_max','y_max']
labels_df.columns = labels_df_columns
labels_df['image_id'] = labels_df['image_id'].apply(lambda x: f"{x:08d}")
# print(labels_df)
labels_df.iloc[:1000]
labels_df.shape

images = []
target_size = (64, 64)

for index, row in labels_df.iterrows():
    img_path = os.path.join(images_dir, f"{row['image_id']}.jpg")
    img = cv2.imread(img_path)
    if img is not None:
        img = cv2.resize(img, target_size)
        images.append(img)
    else:
        print(f"Image not found: {img_path}")

images = np.array(images)

if len(images) == 0:
    print("No images loaded. Please check the image paths.")
else:
    print(f"{len(images)} images loaded successfully.")
```
